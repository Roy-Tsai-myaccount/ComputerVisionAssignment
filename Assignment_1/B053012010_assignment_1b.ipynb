```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy import linalg
%matplotlib inline 
import math

def show_gradient_descent(A, b):
#  enter your code here, and remove the following line (30%)
   
    x = np.array([[-2],[2]])
    
    alpha = 0.02
    max_iter = 1000
    f = np.linalg.norm(A.dot(x) - b)**2
    for k in range(max_iter):
        x1 = x[0]
        x2 = x[1]
        print(k,'(%.4f,%.4f),%.8f'%(x1,x2,f))
              
        x_prev = x


        grad_f = 2*(np.dot(A.T,(np.dot(A,x) - b)))
        x = x - alpha*grad_f
        if(np.linalg.norm(x-x_prev)< 10**(-8)):
            break
        else:
            f =  np.linalg.norm(A.dot(x) - b)**2
  
```
